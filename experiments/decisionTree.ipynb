## ID3(信息增益公式)
$$ IG(S|T)=Entropy(s)-\sum _{value(T)}{\frac{|S_{v}|}{S}}Entropy(S_{v}) $$
其中 S 为全部样本集合，value(T) 是属性 T 所有取值的集合，v 是 T 的其中一个属性值，Sv是 S 中属性 T 的值为 v 的样例集合，|Sv| 为 Sv 中所含样例数。

此时$Entropy(s)=-\frac{8}{17}log(\frac{8}{17})-\frac{9}{17}log(\frac{9}{17})$

当选label为色泽时，此时T为色泽集合，T={‘青绿’，‘乌黑’，‘浅白’}

```python
data[data['色泽']=='青绿']
```

此时$Entropy('青绿')=-\frac{3}{6}log(\frac{3}{6})--\frac{3}{6}log(\frac{3}{6})$

```python
data[data['色泽']=='乌黑']
```

此时$Entropy('乌黑')=-\frac{4}{6}log(\frac{4}{6})--\frac{2}{6}log(\frac{2}{6})$

```python
data[data['色泽']=='浅白']
```

此时$Entropy('青绿')=-\frac{1}{5}log(\frac{1}{5})--\frac{4}{5}log(\frac{4}{5})$

那么$\sum _{value(T='色泽')}{\frac{|S_{v}|}{S}}Entropy(S_{v})=\frac{6}{17}Entropy('青绿')+\frac{6}{17}Entropy('乌黑')+\frac{5}{17}Entropy('浅白')$

所以$IG(S|T='色泽')=Entropy(s)-\sum _{value(T='色泽')}{\frac{|S_{v}|}{S}}Entropy(S_{v})$)

### Very Very Very Brief Version of Decision Treee

```python
import math
import numpy as np
```

```python
def get_ent(df):
    ent = 0
    ctdf = df["好坏"].value_counts()
    for cl in set(list(df["好坏"])):
        prop = float(float(ctdf[cl])/float(sum(ctdf)))
        ent -=prop* math.log(prop)
    return ent
```

```python
class tree:
    def __init__(self,):
        pass
    
    def col_gains(self,df):
        feat_cols = list(set(df.columns)-set(["好坏"]))
        gain_map = list()
        ent0 = get_ent(df)
        if ent0 == 0:
            return list(df["好坏"])[0]
        for col in feat_cols:
            col_gain = ent0
            cts = df[col].value_counts()
            for t in list(set(df[col])):
                col_gain-=(cts[t]/len(df))*get_ent(df[df[col]==t])
            gain_map.append(col_gain)
        final_col = feat_cols[np.array(gain_map).argmax()]
        return dict({final_col:dict((i,self.col_gains(df[df[final_col]==i])) for i in list(set(df[final_col])))})
```

```python
Tree = tree()
```
